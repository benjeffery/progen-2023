# Visualization

SVG-based visualizations of local trees along the genome are easy to produce. Styling the resulting SVG allows a wide range of visualizations, such as 3D effects and even animations. See https://tskit.dev/tutorials/viz.html to see a range of possibilities.

```python
import tskit
ts = tskit.load("data/viz.trees")

css = ".tree .plotbox {transform: skewY(0.8rad)} .x-axis {transform: translateY(80px)}"
sz = (500, 200) # width, height of svg
ts.draw_svg(size=sz, x_scale="treewise", style=css, canvas_size=(sz[0]+10, sz[1]+80))
```
